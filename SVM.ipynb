```python
x = df.drop('spam',axis=1)
y = df['spam']
```

```python
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
x = sc.fit_transform(x)
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.3,random_state=101)
```

```python
from sklearn.svm import SVC
svm = SVC(C=1)
svm.fit(x_train,y_train)
```

```python
pred = svm.predict(x_test)
```

```python
from sklearn import metrics
metrics.accuracy_score(y_test,pred)
```

```python
print(metrics.classification_report(y_test,pred))
```

```python
from sklearn.model_selection import KFold
```

```python
folds = KFold(n_splits=5, shuffle=True,random_state=101)
model = SVC(C=1, random_state=101)
```

```python
from sklearn.model_selection import cross_val_score
```

```python
cv_results = cross_val_score(model, x_train,y_train,cv=folds,scoring='accuracy')
cv_results[:5]
```

```python
cv_results[:5].mean()
```

```python
from sklearn.model_selection import GridSearchCV
```

```python
parameters = {
    'C' : [0.1, 1, 10, 100, 1000]
}
model = SVC(random_state=101)
gd = GridSearchCV(estimator=model,
                 param_grid=parameters,
                 cv=folds,
                 scoring='accuracy',
                 return_train_score=True,
                 verbose=1)

gd.fit(x_train,y_train)
```

```python
gd.best_params_
```

```python
pred = gd.predict(x_test)
```

```python
print(metrics.classification_report(y_test,pred))
```

```python
params = {
    
    'C': [0.1,1,10,100,1000]
}
scores = ['accuracy','precision','recall']
model = SVC(random_state=101)
for score in scores:
    print(f'hyperparameter tuning for {score}')
    
    clf = GridSearchCV(estimator=model,
                      param_grid=parameters,
                      cv=folds,
                      scoring=score,
                      return_train_score=True,
                      verbose=1)
    clf.fit(x_train,y_train)
    print(f'The highest {score} score is {clf.best_score_} at C = {clf.best_params_}')
    print('\n')
```

```python
kernels = ['Polynomial', 'RBF', 'Sigmoid','Linear']

def getClassifier(ktype):
    if ktype == 0:
        return SVC(kernel='poly',degree=8,gamma='auto')
    elif ktype == 1:
        # Radial Basis Function kernal
        return SVC(kernel='rbf', gamma="auto")
    elif ktype == 2:
        # Sigmoid kernal
        return SVC(kernel='sigmoid', gamma="auto")
    elif ktype == 3:
        # Linear kernal
        return SVC(kernel='linear', gamma="auto")
```

```python
from sklearn.metrics import classification_report
for i in range(4):
    svclassifier = getClassifier(i)
    
    svclassifier.fit(x_train,y_train)
    
    y_pred = svclassifier.predict(x_test)
    
    print(f'Evaluation: {i} kernal')
    print(classification_report(y_test,y_pred))
```

```python
param_grid = {
    'C': [0.1,1,10,100],
    'gamma': [1,0.1,0.01, 0.001],
    'kernel': ['rbf', 'poly','sigmoid']
}

grid = GridSearchCV(SVC(),
                   param_grid=param_grid,
                   refit=True,
                   verbose=2,
                   return_train_score=True)

grid.fit(x_train,y_train)
```

```python
best_score = grid.best_score_
best_hyperparams = grid.best_params_
print(f'The best score is {best_score} corresponding to hyperparameters {best_hyperparams}')
```

```python
model = SVC(C=100, gamma=0.001,kernel='rbf')

model.fit(x_train,y_train)

y_pred = model.predict(x_test)

print(f'confusion matrix is {metrics.confusion_matrix(y_test,y_pred)}')
print(f'accuracy: {metrics.accuracy_score(y_test,y_pred)}')
print(f'precision: {metrics.precision_score(y_test,y_pred)}')
print(f'sensicivity/recall: {metrics.recall_score(y_test,y_pred)}')
```
